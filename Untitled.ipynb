```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder

file_path = 'D:\\CSRE\\2. Spring 2024\\IE 506 Machine Learning\\Course Project\\Implementation\\RE-WKLR\\Datasets\\ionosphere.csv'

df = pd.read_csv(file_path, header=None)
df.head()
```

```python
column_names = [f'feature_{i}' for i in range(34)] + ['label']
df.columns = column_names

label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])
# Reverse the encoding
df['label'] = 1 - df['label']  # Now 'g'=0, 'b'=1
```

```python
df.head()
```

```python
X = df.drop(columns=['label']).values
y = df['label'].values
```

```python
(X_train_bal, y_train_bal), (X_train_imb, y_train_imb), (X_test_final, y_test_final) = sampling_preprocessing_pipeline(X, y)
```

```python
import numpy as np
np.shape(y_train_bal)
```

```python
y_train_bal = np.asarray(y_train_bal).flatten()
```

```python
model = WeightedKernelLogisticRegression(sigma=2.5, lambda_=0.07, tau=0.8)

model.fit(X_train_bal, y_train_bal)
pred_probs = model.predict_proba(X_test_final)
pred_labels = model.predict(X_test_final)
```
